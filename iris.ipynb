```python
#Calculating the mean,median and sum of a particular column
sum_data=data["SepalLengthCm"].sum()
mean_data=data["SepalLengthCm"].mean()
median_data=data["SepalLengthCm"].median()
print("Sum:",sum_data, "\nMean:",mean_data,"\nMedian:",median_data)
```

```python
#Extracting minimumand maximum from a column
min_data=data["SepalLengthCm"].min()
max_data=data["SepalLengthCm"].max()
print("Maximum:",max_data, "\nMinimum:",min_data)
```

```python
#Adding a column to the dataset
col=data.columns
print(col)
col=col[1:5]
data1=data[col]
data["total_values"]=data1[col].sum(axis=1)
data
```
